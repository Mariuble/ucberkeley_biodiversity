## Lupinus

```python
bool_array = pollinator1['PLTSP_NAME'].str.lower().str.contains('lupinus', na=False) # length: 1368
lupinus = pollinator1[bool_array]
lupinus.sample(5)
```

```python
pollinator1.groupby('PLTSP_CODE').count()
```

```python
plant_types = pollinator1.groupby('PLTSP_CODE')[['PLTSP_CODE']].value_counts().sort_values().to_frame().reset_index().rename(columns={0:'count'})
plant_types
```

```python
plant_types[(plant_types['count'] > 5) & (plant_types['count'] < 2000)].plot()
```

```python
sns.barplot(data=plant_types[(plant_types['count'] > 200) & (plant_types['count'] < 200000)], x='PLTSP_CODE', y='count')
```

```python
plant_types[plant_types['PLTSP_CODE'] == 'LUPILAXI']
```

### Interactions between Lupinus
